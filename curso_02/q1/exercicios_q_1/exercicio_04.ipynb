```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
import numpy as np

data = np.array([
        [15, 28, 30],
        [410, -2.5, 65],
        [5, 30, 80]
    ])

min_max_scaler = MinMaxScaler()
data_min_max_scaled = min_max_scaler.fit_transform(data)

print(data_min_max_scaled)

plt.figure(figsize=(12, 10))
plt.subplot(2, 1, 2)
plt.scatter(data_min_max_scaled[:, 0], data_min_max_scaled[:, 1], color='green', alpha=0.5)
plt.xlabel('Rooms (Min-Max Scaled)')
plt.ylabel('Age (Min-Max Scaled)')
plt.title('Com Normalização Min-Max Scaling')
plt.show()
```
